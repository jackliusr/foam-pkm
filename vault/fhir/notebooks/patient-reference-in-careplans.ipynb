```polyglot-notebook
#r "nuget:System.Net.Http"
#r "nuget:System.Text.Json"
#r "nuget:Newtonsoft.Json"
#r "nuget:DevLab.JmesPath"


using System.Net.Http;
using System.Text.Json;
using System.Threading.Tasks;
```

```polyglot-notebook
var httpClient = new HttpClient();
var response = await httpClient.GetAsync("https://hapi.fhir.org/baseR4/CarePlan/_search?subject=Patient%2F53254");
JsonElement data  = default;
if (response.IsSuccessStatusCode)
{
    var jsonString = await response.Content.ReadAsStringAsync();
    data = JsonConvert.Deserialize<JsonElement>(jsonString);
    Console.WriteLine(data);
}
else
{
    Console.WriteLine($"Failed to retrieve data: {response.StatusCode}");
}
```
